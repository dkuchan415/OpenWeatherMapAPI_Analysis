### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Configure Gmaps

gmaps.configure(api_key=g_key)

#Create coordinates and humidity list

locations = vaca_df[['Lat', 'Lng']].astype(float)

humidity = vaca_df['Humidity'].astype(float)

#Create for loop to add Lat, Lng, and humidty for each city

# cities = vaca_df['City']
# city_name = []
# lat = []
# lng = []
# humidity = []

# for city in cities:
#     city_name.append(vaca_df['City'])
#     lat.append(vaca_df['Lat'])
#     lng.append(vaca_df['Lng'])
#     humidity.append(vaca_df['Humidity'])

figure_layout = {
    'width': '400px',
    'height': '300px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)

# Assign the marker layer to a variable

heat_layer=gmaps.heatmap_layer(locations, humidity, dissipating=True, max_intensity=100, point_radius=10)

# Add the layer to the map
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
#cut unneeded columns
hotel_df = vaca_df.dropna(how='any')

hotel_df = hotel_df.loc[(hotel_df['Max Temp'] < 85) &
                       (hotel_df['Max Temp'] > 75) &
                       (hotel_df['Cloudiness'] < 15) &
                       (hotel_df['Wind Speed'] < 5)]
len(hotel_df)
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#Day 3 activity 4, iterate through df and search for hotel by coordinates

hotel_df['Hotel Name'] = ""

#Add goople places base url

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

#Set base parameters

params = {"radius": "5000",
        "type": "lodging",
        "key": g_key,
}
for index, row in hotel_df.iterrows():
    lat_lng = f"{row['Lat']},{row['Lng']}"
    params['location'] = lat_lng
    response = requests.get(base_url, params=params).json()
    print(json.dumps(response, indent=4))
#plot markers = name of hotel- use jupyter gmaps
    hotel_df.loc[index, 'Hotel Name'] = response['results'][0]['name']
# change map options and adjusting

#use weighted heat map
```
